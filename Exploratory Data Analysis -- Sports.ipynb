### Data Processing

```python
# All city with missing values.
matches[matches.city.isnull()] [['city','venue']]
```

```python
matches.city=matches.city.fillna('Dubai')
```

```python
matches[(matches.umpire1.isnull()) | (matches.umpire2.isnull())]
```

```python
# Umpire3 column has close to 92% missing values. hence dropping that column
matches = matches.drop('umpire3', axis = 1)
```

```python
city_venue = matches.groupby(['city','venue']).count()['season']
city_venue_df = pd.DataFrame(city_venue)
city_venue_df
```

```python
matches['city'] = matches['city'].replace(['Bengaluru'],'Bangalore')
matches['city'] = matches['city'].replace(['Mohali'],'Chandigarh')
matches['venue'] = matches['venue'].replace(['M Chinnaswamy Stadium'],'M. Chinnaswamy Stadium')
matches['venue'] = matches['venue'].replace(['IS Bindra Stadium'],'PCA IS Bindra Stadium, Mohali')
matches['venue'] = matches['venue'].replace(['Punjab Cricket Association Stadium, Mohali'],'PCA IS Bindra Stadium, Mohali')
matches['venue'] = matches['venue'].replace(['Punjab Cricket Association IS Bindra Stadium, Mohali'],'PCA IS Bindra Stadium, Mohali')
matches['venue'] = matches['venue'].replace(['M. A. Chidambaram Stadium'],'M. A. Chidambaram Stadium, Chepauk')
matches['venue'] = matches['venue'].replace(['MA Chidambaram Stadium, Chepauk'],'M. A. Chidambaram Stadium, Chepauk')
matches['venue'] = matches['venue'].replace(['Feroz Shah Kotla'],'Feroz Shah Kotla Ground')
matches['venue'] = matches['venue'].replace(['Rajiv Gandhi International Stadium, Uppal'],'Rajiv Gandhi Intl. Cricket Stadium')
matches['venue'] = matches['venue'].replace(['Dr. Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium'],'ACA-VDCA Stadium')
matches['venue'] = matches['venue'].replace(['Subrata Roy Sahara Stadium'],'Maharashtra Cricket Association Stadium')
```

```python
city_venue = matches.groupby(['city','venue']).count()['season']
city_venue_df = pd.DataFrame(city_venue)
city_venue_df
```

## Data visualization

```python
#Plotting venues along with cities 
v = pd.crosstab(matches['city'],matches['venue'])


#Adding a column by summing each columns
v['count'] = v.sum(axis = 'columns')
#We will just keep last column = 'count'
b = v['count']

#Plotting
plt.figure(figsize = (20,10))
b.plot(kind = 'bar')
plt.title("Number of matches in different cities", fontsize = 25, fontweight = 'bold', color='green')
plt.xlabel("City", size = 30, color='blue')
plt.ylabel("Frequency", size = 30, color='green')
plt.xticks(size = 20, color='black')
plt.yticks(size = 20, color='black')
plt.show()
```

```python
# Plot to visualise the no. of matches held in each city
plt.figure(figsize = (15,15))
sns.countplot(y='venue', data=matches, color='green')
plt.title('No. of matches held in each venue', fontsize = 50, fontweight = 'bold', color='green')
plt.xlabel('No. of matches held', size = 50, color='blue')
plt.ylabel('Venues', size =50, color='blue')
plt.xticks(size =20, color='black')
plt.yticks(size =20, color='black')
plt.xlim([0,120])
plt.show()
```

```python
matches.venue.value_counts().sort_values(ascending = True).tail(10).plot(kind = 'barh',figsize=(20,20), fontsize=15, color='green')
plt.title("Venue which has hosted most number of IPL matches",fontsize=35,fontweight="bold", color='green')
plt.ylabel("Venue", size = 25, color='blue')
plt.xlabel("Frequency", size = 25, color='blue')
plt.xticks(size = 20, color='black')
plt.yticks(size = 20, color='black')
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.countplot('season', data = matches, palette=sns.color_palette())
plt.title("Number of matches played each season",fontsize=30, color='green')
plt.ylabel("Count", size = 25, color='blue')
plt.xlabel("Season", size = 25, color='blue')
plt.xticks(size = 20, color='black')
plt.yticks(size = 20, color='black')
plt.show()
```

```python
matches.groupby('season')['team1'].nunique().plot(kind = 'bar', figsize=(15,10))
plt.title("Number of teams participated each season ",fontsize=30, color='green')
plt.ylabel("Count of teams", size = 25, color='blue')
plt.xlabel("Season", size = 25, color='blue')
plt.xticks(size = 15, color='black')
plt.yticks(size = 15, color='black')
plt.show()
```

```python
# Plot to visualise the no. of matches won by each team
data=matches['winner'].value_counts()
plt.figure(figsize = (15,10))
sns.barplot(y=data.index, x=data, orient='h', palette=sns.color_palette())
plt.title('No. of matches won by each team',fontsize=35,fontweight="bold", color='green')
plt.xlabel("\nNo. of matches won",size = 25, color='blue')
plt.ylabel('Teams',size = 25, color='blue')
plt.xticks(size = 15, color='black')
plt.yticks(size = 15, color='black')
plt.show()
```

```python
#creating a dataframe with season and winner columns
winning_teams = matches[['season','winner']]
#dictionaries to get winners to each season
winners_team = {}
for i in sorted(winning_teams.season.unique()):
    winners_team[i] = winning_teams[winning_teams.season == i]['winner'].tail(1).values[0]
    
winners_of_IPL = pd.Series(winners_team)
winners_of_IPL = pd.DataFrame(winners_of_IPL, columns=['team'])

#plotting
winners_of_IPL['team'].value_counts().plot(kind = 'barh', figsize = (15,10), color = 'darkblue')
plt.title("Winners of IPL across 12 seasons",fontsize=30,color='green')
plt.ylabel("Teams", size = 25,color='blue')
plt.xlabel("Frequency", size = 25,color='blue')
plt.xticks(size = 20, color='black')
plt.yticks(size = 20, color='black')
plt.show()
```

```python
#obtain toss winners in the match
winner = matches[matches['toss_winner']==matches['winner']]
labels=['Yes','No']
#pie plot
plt.pie([len(winner),(577-len(winner))],labels=labels,autopct='%1.1f%%',radius=2, 
        shadow=True, startangle=90,textprops=dict(color="w"))
plot = plt.gcf()
plt.title("Toss winners' precentage of success\n\n\n\n\n", color="w")
plt.show()
```

```python
# How many times did each team win the toss?
# Plot to visualise the no. of tosses won by each team
toss=matches['toss_winner'].value_counts()
fig, ax=plt.subplots()
ax.set_title('No. of tosses won by each team\n',size = 30, color='green')
sns.barplot(y=toss.index, x=toss, orient='h', palette=sns.color_palette())
plt.xlabel('\nNo. of tosses won',size = 20,color='blue')
plt.ylabel('Teams',size = 20,color='blue')
plt.xticks(size = 10, color='black')
plt.yticks(size = 10, color='black')
plt.show()
```

```python
merge['toss_winner_is_winner'] = 'no'
merge['toss_winner_is_winner'][merge.toss_winner == merge.winner] = 'yes'
plt.figure(figsize=(12,6))
sns.countplot(x='toss_winner', hue='toss_winner_is_winner', data=merge)
plt.xlabel("Teams", size=20, color='orange')
plt.ylabel("Count", size=20, color='orange')
plt.xticks(size = 10,color='black',rotation='vertical')
plt.yticks(size = 10,color='black')
plt.show()
```

```python
merge.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'blue',
                                                                                            figsize = (15,10))
plt.title("Top Run Getters of IPL", fontsize = 20, fontweight = 'bold', color='green')
plt.xlabel("Batsmen", size = 25, color='blue')
plt.ylabel("Total Runs Scored", size = 25, color='blue')
plt.xticks(size = 15, color='black',rotation='vertical')
plt.yticks(size = 15, color='black')
plt.show()
```

```python
no_of_balls = pd.DataFrame(merge.groupby('batsman')['ball'].count()) #total number of matches played by each batsman
runs = pd.DataFrame(merge.groupby('batsman')['batsman_runs'].sum()) #total runs of each batsman
seasons = pd.DataFrame(merge.groupby('batsman')['season'].nunique()) #season = 1 implies played only 1 season

batsman_strike_rate = pd.DataFrame({'balls':no_of_balls['ball'],'run':runs['batsman_runs'],'season':seasons['season']})
batsman_strike_rate.reset_index(inplace = True)

batsman_strike_rate['strike_rate'] = batsman_strike_rate['run']/batsman_strike_rate['balls']*100
highest_strike_rate = batsman_strike_rate[batsman_strike_rate.season.isin([10,11,12])][['season','batsman','strike_rate']].sort_values(by = 'strike_rate',
                                                                                                            ascending = False)

highest_strike_rate.head(10)
```

```python
plt.figure(figsize = (10,10))
sns.barplot(x='batsman', y='strike_rate', data = highest_strike_rate.head(10), hue = 'season')
plt.title("Highest strike rates in IPL",fontsize= 30,color='green')
plt.xlabel("Player", size = 25,color='blue')
plt.ylabel("Strike Rate", size = 25,color='blue')
plt.xticks(size = 15,color='black',rotation='vertical')
plt.yticks(size = 15,color='black')
plt.show()
```

Batsman with most number of boundaries

```python
temp_df = deliveries.groupby('batsman')['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
temp_df = temp_df.iloc[:10,:]

labels = np.array(temp_df['batsman'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots()
rects = ax.bar(ind, np.array(temp_df['batsman_runs']), width=width, color='green')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count", size=20, color='blue')
ax.set_title("Batsman with most number of boundaries", size=30, color='green')
plt.xticks(size = 10,color='black',rotation='vertical')
plt.yticks(size = 10,color='black')
plt.show()
```
